```python
import pyautogui
import time

#pyautogui - biblioteca para automatizar tarefas
#Principais comandos:
#pyautogui.click -> clicar com mouse
#pyautogui.write -> escreve um texto
#pyautogui.press -> aperta uma tecla
#pyautogui.hotkey -> apertar uma combinação de teclas

pyautogui.PAUSE = 1
#pausa de 1 segunda antes de executar o próximo comando

#Passo 1: Acessar o sistema da empresa
pyautogui.hotkey("ctrl", "t")
pyautogui.write("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")
pyautogui.press("enter")

time.sleep(5)
#espera de 5 segundo neste intervalo de códigos

```

```python
#Passo 2: Fazer login no sistema
pyautogui.click(x=686, y=402)

pyautogui.write("meu login")
pyautogui.click(x=693, y=483)
pyautogui.write("minha senha")
#pyautogui.press("enter")
pyautogui.click(x=722, y=560, clicks=2)
#button="left"
time.sleep(5)
```

```python
#Passo 3: Baixar a base de dados
pyautogui.click(x=536, y=355)
pyautogui.click(x=590, y=630)
```

```python
import pandas as pd

#Passo 4: Calcular os indicadores
#importar a base de dados
table = pd.read_csv(r"/home/amanda/Downloads/Compras.csv", sep=";")
display(table)

#calculo dos indicadores
#total gasto -> somar a coluna valor final [] -> seleciona a coluna
total_gasto = table["ValorFinal"].sum()

#quantidade -> somar a coluna quantidade
quantidade = table["Quantidade"].sum()

#preço medio -> somar total gasto / quantidade
preco_medio = total_gasto / quantidade

print(total_gasto)
print(quantidade)
print(preco_medio)

```
